```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Data hewan
data = {
    'Hewan': ['Kuda', 'Manusia', 'Kambing', 'Kucing', 'Tikus'],
    'Massa (kg)': [700, 70, 55, 15, 0.4],
    'Laju Metabolisme (kkal/hari)': [11760, 1700, 1450, 720, 30]
}

# Mengambil logaritma natural
massa_log = np.log(data['Massa (kg)'])
metabolisme_log = np.log(data['Laju Metabolisme (kkal/hari)'])

# Membuat array X dan Y untuk regresi linier
X = massa_log
Y = metabolisme_log

# Melakukan regresi linier
slope, intercept, r_value, p_value, std_err = linregress(X, Y)

print(f"Persamaan Power Law: log(Laju Metabolisme) = {slope:.3f} * log(Massa) + {intercept:.3f}")

# Plotting data dan garis regresi
plt.scatter(massa_log, metabolisme_log)
plt.plot(massa_log, slope * massa_log + intercept, color='red')
plt.xlabel('log(Massa)')
plt.ylabel('log(Laju Metabolisme)')
plt.title('Hubungan Power Law Antara Massa dan Laju Metabolisme')
plt.show()
```

```python
#Ghifari Raffasyah (082111733002)
#UTS Komputasi Biomedis T1
import numpy as np

# Data percobaan
data = {
    'H': [100, 80, 140, 90],
    'K': [125, 110, 80, 104.8],
    'L': [125, 120, 120, 60],
    'M': [62.5, 25, 100, 137.33],
    'F': [6625, 5290, 7300, 6539]
}

# Membuat matriks koefisien dan vektor hasil
A = []
B = []

for i in range(4):
    A.append([data['H'][i], data['K'][i], data['L'][i], data['M'][i]])
    B.append(data['F'][i])

# Mengonversi ke format numpy array
A = np.array(A)
B = np.array(B)

# Menyelesaikan sistem persamaan linier Ax = B
concentrations = np.linalg.solve(A, B)

# Menampilkan hasil konsentrasi
y, w, x, z = concentrations
print(f"Konsentrasi obat dalam setiap aliran:")
print(f"y (fase organik H): {y:.4f} g/L")
print(f"w (fase organik K): {w:.4f} g/L")
print(f"x (fase berair L): {x:.4f} g/L")
print(f"z (fase berair M): {z:.4f} g/L")
```
